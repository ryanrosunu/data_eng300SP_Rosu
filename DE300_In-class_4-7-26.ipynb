```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load and save a copy of the earthquake dataset
earthquake = pd.read_csv('https://raw.githubusercontent.com/mosesyhc/de300-2026wi/refs/heads/main/datasets/Canadian-Earthquakes-2010-2019.csv')
```

```python
earthquake.head()
earthquake.columns
```

```python
sns.histplot(data=earthquake, x='magnitude', binwidth=0.2)
plt.xlabel("Magnitude")
plt.ylabel("Count")
plt.title("Magnitude Distribution")
plt.show()
```

```python
sns.countplot(data=earthquake, x='magnitude_codelist')
plt.xlabel("Category")
plt.ylabel("Count")
plt.title("Magnitude Categories")
plt.show()
```

```python
sns.scatterplot(data=earthquake, x='magnitude', y='depth')
plt.xlabel("Magnitude")
plt.ylabel("Depth (km)")
plt.title("Magnitude vs. Depth")
plt.show()
```
